Question 1:
Write a python program to display IMDB’s Top rated 100 Indian movies’ data
https://www.imdb.com/list/ls056092300/  (i.e. name, rating, year ofrelease) and make data frame.

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
page = requests.get('https://www.imdb.com/list/ls056092300/')
page
```

```python
soup = BeautifulSoup(page.content)
soup
```

```python
movie_items = soup.find_all('div', class_="lister-item-content")
```

```python
names = []
ratings = []
years = []

for movie in movie_items:
    name = movie.find('a').text.strip()
    names.append(name)
    
    
    rating = movie.find('span', class_="ipl-rating-star__rating").text.strip()
    ratings.append(float(rating))

    year = movie.find('span', class_="lister-item-year").text.strip()
    years.append(year)

data = {'Name': names, 'Rating': ratings, 'Year': years}
df = pd.DataFrame(data)
print(df)

```

a) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.

Question 4):
Write a python program to scrape details of all the posts from https://www.patreon.com/coreyms .Scrape the
heading, date, content and the likes for the video from the link for the youtube video from the post

```python
import requests
from bs4 import BeautifulSoup


def scrape_patreon_posts(url):
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        
        posts = []
        post_elements = soup.find_all('div', class_='post')
        
        for post_element in post_elements:
            heading = post_element.find('h3', class_='post__title').text.strip()
            date = post_element.find('time', class_='post__date').text.strip()
            content = post_element.find('div', class_='post__content').text.strip()
            likes = post_element.find('div', class_='post__stats').text.strip()
            
            youtube_link = post_element.find('div', class_='post__content').find('a', href=lambda href: href and "youtube" in href.lower())
            youtube_url = youtube_link['href'] if youtube_link else None
            
            posts.append({'Heading': heading, 'Date': date, 'Content': content, 'Likes': likes, 'YouTube URL': youtube_url})
        
        return posts
    else:
        print("Failed to retrieve page.")
        return None


if __name__ == "__main__":
    url = "https://www.patreon.com/coreyms"
    posts = scrape_patreon_posts(url)
    
    if posts:
        for idx, post in enumerate(posts, start=1):
            print(f"Post {idx}:")
            for key, value in post.items():
                print(f"{key}: {value}")
            print("=" * 50)
    else:
        print("No posts found.")
```

Question 5)
Write a python program to scrape house details from mentioned URL. It should include house title, location,
area, EMI and price from https://www.nobroker.in/ .Enter three localities which are Indira Nagar, Jayanagar,
Rajaji Nagar.


```python
import requests
from bs4 import BeautifulSoup

def scrape_house_details(localities):
    base_url = "https://www.nobroker.in/property/sale/"
    house_details = []
    
    for locality in localities:
        url = base_url + locality.replace(" ", "-")
        response = requests.get(url)
        
        if response.status_code == 200:
            soup = BeautifulSoup(response.content, 'html.parser')
            houses = soup.find_all('div', class_='card')
            
            for house in houses:
                title = house.find('h2', class_='card-title').text.strip()
                location = house.find('div', class_='card-subtitle').text.strip()
                area = house.find('div', class_='card-size').text.strip()
                emi = house.find('div', class_='card-emi').text.strip()
                price = house.find('div', class_='card-price').text.strip()
                
                house_details.append({'Title': title, 'Location': location, 'Area': area, 'EMI': emi, 'Price': price})
                
        else:
            print(f"Failed to retrieve data for {locality}.")
            
    return house_details

def print_house_details(house_details):
    for idx, house in enumerate(house_details, start=1):
        print(f"House {idx}:")
        for key, value in house.items():
            print(f"{key}: {value}")
        print("=" * 50)
        
if __name__ == "__main__":
    localities = ['Indira Nager', 'Jayanagar', 'Rajaji Nagar']
    house_details = scrape_house_details(localities)
    print_house_details(house_details)
```

Question 6)
Write a python program to scrape first 10 product details which include product name , price , Image URL from
https://www.bewakoof.com/bestseller?sort=popular 

```python
import requests
from bs4 import BeautifulSoup

def scrape_bewakoof_bestsellers(url):
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        
        products = []
        product_cards = soup.find_all('div', class_='productCardBox')
        
        for card in product_cards[:10]:
            product_name = card.find('div', class_='product-title').text.strip()
            price = card.find('div', class_='product-price').text.strip()
            image_url = card.find('img')['src']
            
            products.append({'Product Name': product_name, 'Price': price, 'Image URL': image_url})
        
        return products
    else:
        print("Failed to retrieve page.")
        return None

if __name__ == "__main__":
    url = "https://www.bewakoof.com/women-tshirts?ga_q=tshirts"
    products = scrape_bewakoof_bestsellers(url)
    
    if products:
        for idx, product in enumerate(products, start=1):
            print(f"Product {idx}:")
            for key, value in product.items():
                print(f"{key}: {value}")
            print("=" * 50)
    else:
        print("No products found.")
```
